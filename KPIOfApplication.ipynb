```python
from cryptography.fernet import Fernet
from datetime import datetime

```

```python
message = "apptesting".encode()
key = Fernet.generate_key()
f = Fernet(key)
```

```python
encrypted = f.encrypt(message)  # Encrypt the bytes. The returning object is of type bytes
decrypted = f.decrypt(encrypted)  # Decrypt the bytes. The returning object is of type bytes
```

```python
import numpy as np
```

```python
encrypted = f.encrypt(message)  # Encrypt the bytes. The returning object is of type bytes
decrypted = f.decrypt(encrypted)  # Decrypt the bytes. The returning object is of type bytes
usage_start_time = datetime.now()
enctime =np.zeros(10)
inputlen =np.zeros(10)
outputlen =np.zeros(10)
dectime =np.zeros(10)
for i in range(10):
  message = message*(i+1)
  start_time = datetime.now()
  encrypted = f.encrypt(message) 
  end_time = datetime.now()
  time_difference = (end_time - start_time).total_seconds() * 10**3
  enctime[i] = time_difference
  inputlen[i] = len(message)
  outputlen[i] = len(encrypted)
  print("Execution time for encryption of text with length:", inputlen[i], "is:", time_difference, "ms") 
  
  start_time = datetime.now()
  decrypted = f.decrypt(encrypted)
  end_time = datetime.now()
  time_difference = (end_time - start_time).total_seconds() * 10**3
  dectime[i] = time_difference
  print("Execution time for decryption of text with length:", outputlen[i], "is:", time_difference, "ms")
  usage_end_time = datetime.now()
  usage_time_difference = (usage_end_time - usage_start_time).total_seconds() * 10**3
```

```python
avgInput = np.mean(inputlen)
avgOutput = np.mean(outputlen)
avgEncTime = np.mean(enctime)
avgDecTime = np.mean(dectime)
```

```python
print("#### KPI Matrix of Encryption API ####")
print("Usage Count in:", usage_time_difference, "ms is:", 10)
print("Average Request size:", avgInput)
print("Average Response size:", avgOutput)
print("Average Latency:", avgEncTime)
print('\n')
print("#### KPI Matrix of Encryption API ####")
print("Usage Count in:", usage_time_difference, "ms is:", 10)
print("Average Request size:", avgOutput)
print("Average Response size:", avgInput)
print("Average Latency:", avgDecTime)
```
