```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
```

```python
models = [XGBClassifier(), LGBMClassifier(), DecisionTreeClassifier(), RandomForestClassifier(), AdaBoostClassifier(), MLPClassifier(), KNeighborsClassifier()]
model_names = ['XGBClassifier()', 'LGBMClassifier()', 'DecisionTreeClassifier()', 'RandomForestClassifier()', 'AdaBoostClassifier()', 'MLPClassifier()', 'KNeighborsClassifier()']
```

```python
#df.info()
df_new = df.iloc[:, [1,2,3,5,6,13,20,21,23,25,26,27,28,29,42,43]]
df_new.info()
```

```python
df_new.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_new.drop(['is_claim'], axis=1), df_new['is_claim'], test_size=0.3, shuffle=True)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
testmodel = KNeighborsClassifier()
testmodel.fit(X_train, y_train)
testmodel.score(X_test, y_test)
```

```python
for i, model in enumerate(models):
    model.fit(X_train, y_train)
    print(f'{model_names[i]}: Accuracy: {model.score(X_test, y_test)}')
    print()
```
